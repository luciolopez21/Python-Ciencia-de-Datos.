
**Ejercicio 3:**

Crear un programa donde el usuario ingrese la cantidad de filas y columnas que tendra una tabla de datos. Luego el programa pedira ingresar los datos de la tabla fila por fila. Todos los datos serán numéricos.

Mostrar la tabla ingresada en formato LISTA de Python, y mostrar la misma tabla en formato array de Numpy.

Solicitar al usuario que ingrese las posiciones de dos filas y realice la suma  de las mismas. Mostrar este vector resultado.

```python
import numpy as np

# Pedir al usuario la cantidad de filas y columnas de la tabla
num_filas = int(input("Ingrese la cantidad de filas: "))
num_columnas = int(input("Ingrese la cantidad de columnas: "))

# Crear una lista vacía para almacenar los datos de la tabla
tabla_datos = []

# Pedir al usuario que ingrese los datos fila por fila
for i in range(num_filas):
    fila = []
    for j in range(num_columnas):
        dato = float(input(f"Ingrese el dato para la fila {i+1}, columna {j+1}: "))
        fila.append(dato)
    tabla_datos.append(fila)

# Convertir la lista en un array de Numpy
array_numpy = np.array(tabla_datos)

# Mostrar la tabla en formato de lista de Python
print("Tabla ingresada (formato lista):")
for fila in tabla_datos:
    print(fila)

# Mostrar la tabla en formato de array de Numpy
print("\nTabla ingresada (formato array de Numpy):")
print(array_numpy)

# Solicitar al usuario que ingrese las posiciones de dos filas para la suma
fila1 = int(input("Ingrese la posición de la primera fila para la suma (1-{0}): ".format(num_filas))) - 1
fila2 = int(input("Ingrese la posición de la segunda fila para la suma (1-{0}): ".format(num_filas))) - 1

# Realizar la suma de las filas seleccionadas
suma_filas = array_numpy[fila1] + array_numpy[fila2]

# Mostrar el vector resultado de la suma de filas
print("\nVector resultado de la suma de filas:")
print(suma_filas)
```

A continuación se muestran los valores de  los siguientes productos:

['arroz', 'harina','fideo','yerba','azucar']=[145.6 , 100 , 89.90 , 700 , 95]

Los valores de estos productos son aproximados de hace dos meses, debido a la inflación y alza de los precios, se vieron afectados de la siguiente manera:
*   Producto arroz , harina, azucar duplicaron su precio
*   Productos restantes incrementaron en un 75% su precio

Mostrar los datos en forma de vector y actualizar sus precios, de manera que se pueda comparar ambos vectores.

```python
import numpy as np

# Valores iniciales de los productos
productos = ['arroz', 'harina', 'fideo', 'yerba', 'azucar']
valores_iniciales = [145.6, 100, 89.90, 700, 95]

# Actualizar los precios según las indicaciones
valores_actualizados = []
for i, producto in enumerate(productos):
    if producto in ['arroz', 'harina', 'azucar']:
        nuevo_precio = valores_iniciales[i] * 2
    else:
        nuevo_precio = valores_iniciales[i] * 1.75
    valores_actualizados.append(nuevo_precio)

# Convertir las listas en vectores de Numpy
vector_inicial = np.array(valores_iniciales)
vector_actualizado = np.array(valores_actualizados)

# Mostrar los vectores
print("Valores iniciales:")
print(vector_inicial)

print("\nValores actualizados:")
print(vector_actualizado)
```
