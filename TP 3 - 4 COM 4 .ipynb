##***EJERCICIO 1***

Simular el lanzamiento de un dado eqilibrado de 6 caras.


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación del lanzamiento de un dado equilibrado de 6 caras
num_lanzamientos = 100
resultados = np.random.randint(1, 7, size=num_lanzamientos)

# Cálculo de las probabilidades aproximadas
prob_2 = np.count_nonzero(resultados == 2) / num_lanzamientos
prob_3 = np.count_nonzero(resultados == 3) / num_lanzamientos
prob_5 = np.count_nonzero(resultados == 5) / num_lanzamientos

# Impresión de las probabilidades aproximadas
print(f"Probabilidad de obtener el número 2: {prob_2:.4f}")
print(f"Probabilidad de obtener el número 3: {prob_3:.4f}")
print(f"Probabilidad de obtener el número 5: {prob_5:.4f}")

# Creación del histograma de las probabilidades
numeros = ['2', '3', '5']
probabilidades = [prob_2, prob_3, prob_5]

plt.bar(numeros, probabilidades, color=['blue', 'green', 'red'])
plt.xlabel('Número en el dado')
plt.ylabel('Probabilidad')
plt.title('Histograma de las probabilidades de obtener los números 2, 3 y 5')
plt.show()

```

## EJERCICIO 5  

Solicitar al usuario que ingrese la cantidad N de datos a cargar. Luego generar de manera aleatoria los siguientes datos:

* SEXO= 'MASCULINO', 'FEMENINO','NO ESPECIFICA'
* EDAD= entre 28 y 60 años
* ALTURA= entre un rango de 165.0 mts hasta 210 mts
* PESO= entre un rango de 50  y 110 kg
* HORAS DE DESCANSO diario= un número decimal
* HORAS DE TRABAJO/ ESTUDIO diario= un número decimal
* HORAS DE EJERCICIOS diario= un número decimal

REalizar las siguientes indicaciones y extraer concluisiones sobre la información y la pooblación muestreada.

1.   Generar un array de dimensión 2 (matriz), con todos los datos ordenados, permitiendo visualizarlos como tabla.
2.   Suponiendo que la distribución de probabilidad de los datos muestrados de las N personas, se distribuyen en forma NORMAL. Encontrar la media en los datos de altura, peso, y horas de descanso, trabajo y ejercicios.
3.  Graficar la muestra o distribución de los datos  de pesos , y alturas en graficas separadas. Indicar el grado de variabilidad de sus datos
4.  Plotear de manera conjunta y encontrar si presentan correlación los datos de peso vs altura,
5. Plotear de manera conjunta los datos de horas de descanso, estudio y ejercicio



```python
import numpy as np
import matplotlib.pyplot as plt

# Solicitar al usuario la cantidad de datos a cargar
N = int(input("Ingrese la cantidad de datos a cargar: "))

# Generar datos aleatorios
SEXO = np.random.choice(['MASCULINO', 'FEMENINO', 'NO ESPECIFICA'], size=N)
EDAD = np.random.randint(28, 61, size=N)
ALTURA = np.random.uniform(165.0, 210.0, size=N)
PESO = np.random.uniform(50, 110, size=N)
DESCANSO = np.random.uniform(4, 10, size=N)
TRABAJO_ESTUDIO = np.random.uniform(4, 12, size=N)
EJERCICIOS = np.random.uniform(0, 3, size=N)

# Crear matriz de datos
datos = np.array([SEXO, EDAD, ALTURA, PESO, DESCANSO, TRABAJO_ESTUDIO, EJERCICIOS]).T

# Imprimir matriz como tabla
headers = ['SEXO', 'EDAD', 'ALTURA', 'PESO', 'DESCANSO', 'TRABAJO/ESTUDIO', 'EJERCICIOS']
print('\t'.join(headers))
for row in datos:
    print('\t'.join(map(str, row)))

# Calcular la media de altura, peso, horas de descanso, trabajo y ejercicios
media_altura = np.mean(ALTURA)
media_peso = np.mean(PESO)
media_descanso = np.mean(DESCANSO)
media_trabajo_estudio = np.mean(TRABAJO_ESTUDIO)
media_ejercicios = np.mean(EJERCICIOS)

print(f"Media de ALTURA: {media_altura:.2f}")
print(f"Media de PESO: {media_peso:.2f}")
print(f"Media de DESCANSO: {media_descanso:.2f}")
print(f"Media de TRABAJO/ESTUDIO: {media_trabajo_estudio:.2f}")
print(f"Media de EJERCICIOS: {media_ejercicios:.2f}")

# Graficar la distribución de pesos y alturas
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(PESO, bins=20, color='blue', alpha=0.7)
plt.xlabel('Peso (kg)')
plt.ylabel('Frecuencia')
plt.title('Distribución de Pesos')

plt.subplot(1, 2, 2)
plt.hist(ALTURA, bins=20, color='green', alpha=0.7)
plt.xlabel('Altura (cm)')
plt.ylabel('Frecuencia')
plt.title('Distribución de Alturas')

plt.tight_layout()
plt.show()

# Graficar correlación entre peso y altura
plt.scatter(PESO, ALTURA, color='purple', alpha=0.5)
plt.xlabel('Peso (kg)')
plt.ylabel('Altura (cm)')
plt.title('Correlación entre Peso y Altura')
plt.show()

# Graficar horas de descanso, trabajo y ejercicios
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.hist(DESCANSO, bins=20, color='blue', alpha=0.7)
plt.xlabel('Horas de Descanso')
plt.ylabel('Frecuencia')
plt.title('Distribución de Horas de Descanso')

plt.subplot(1, 3, 2)
plt.hist(TRABAJO_ESTUDIO, bins=20, color='green', alpha=0.7)
plt.xlabel('Horas de Trabajo/Estudio')
plt.ylabel('Frecuencia')
plt.title('Distribución de Horas de Trabajo/Estudio')

plt.subplot(1, 3, 3)
plt.hist(EJERCICIOS, bins=20, color='purple', alpha=0.7)
plt.xlabel('Horas de Ejercicios')
plt.ylabel('Frecuencia')
plt.title('Distribución de Horas de Ejercicios')

plt.tight_layout()
plt.show()
```
